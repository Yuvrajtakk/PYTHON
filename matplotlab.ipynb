```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

# **line plot**

```python
x = [1, 2, 3, 4, 5]
y = [2, 3,5, 7, 11]
plt.plot(x,y)
plt.xlabel('Roll number')
plt.ylabel('Marks')
plt.title('Line plot')
plt.show()
```

# **Bar Graph**

```python
catogories = ['A', 'B', 'C', 'D', 'E' ]
values = [2, 3, 5, 7, 11]

plt.bar(catogories, values)
plt.xlabel('Roll number')
plt.ylabel('Marks')
plt.title('Bar Graph')
plt.show()
```

# **Histograms**

```python
new = np.random.randn(500)

plt.hist(new, bins=5)
plt.xlabel('Roll number')
plt.ylabel('Marks')
plt.title('Histogram')
plt.show()
```

# **Pie Chart**

```python
catogories = ['Friuts', 'vegetables', 'Clothes', 'Shoes' ]
expenses = [500, 300, 1000, 700]

plt.pie(expenses, labels= catogories, autopct= '%1.1f%%')
plt.title("Monthly expense for June")
plt.show()
```
