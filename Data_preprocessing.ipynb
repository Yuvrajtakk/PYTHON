```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", np.nan, "Eve", "Frank", "Grace", "Heidi", "Ivan", "Judy"],
    "Age": [25, 32, np.nan, 45, 29, 34, np.nan, 41, 30, 27],
    "Gender": ["Female", "Male", "Male", "male", np.nan, "female", "Female", "Male", "Male", "Female"],
    "Income": [50000, 60000, 55000, 45000, np.nan, 70000, 52000, 48000, 62000, 50000],
    "City": ["New York", "Los Angeles", "New York", "Chicago", "Los Angeles", np.nan, "Chicago", "New York", "Chicago", "Chicago"],
    "Has_Debt": ["Yes", "No", "Yes", "No", "Yes", "Yes", "No", np.nan, "Yes", "No"],
    "Debt_Amount": [15000, 0, 12000, 0, 18000, 20000, 0, 14000, np.nan, 0]
}
```

```python
df = pd.DataFrame(data)
```

```python
df['Name'] = df['Name'].fillna('Unknown')
```

```python
df
```

```python
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df
```

```python
df['Income'] = df['Income'].fillna(df['Income'].mean())
df
```

```python
#gender
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df
```

```python
#city

df['City'] = df['City'].ffill()
df
```

```python
#hasdebt

df['Has_Debt'] = df['Has_Debt'].fillna('Yes')
df
```

```python
#debt amount

df['Debt_Amount'] = df['Debt_Amount'].fillna('0')
df
```
